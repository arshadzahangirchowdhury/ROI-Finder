```python
def check_clustering_class(dataframe, class_var_string = 'Class', verbosity=False):
    '''
    inputs
    
    dataframe: pandas dataframe, must contain a column called 'Class' containing, kmeans.labels_ and 'K' column consisting of max amount of potassium values
    class_var_string:str, name of the pandas column containing the label values
    returns: dataframe is flipped, 0's are returned as 1 and 1's would be returned as 0's in the Class column of the dataframe to keep consistency.
    
    '''
    if verbosity != False:
        print('clustering algorithm assigned class to max K cell:', dataframe.iloc[dataframe['K'].idxmax()]['Class'])
        print('High K cell should be 1. Is it?', dataframe.iloc[dataframe['K'].idxmax()]['Class'] ==1 )

    
    if dataframe.iloc[dataframe['K'].idxmax()][class_var_string] != 1:
        dataframe.loc[dataframe[class_var_string] == 1, class_var_string] = -99
        dataframe.loc[dataframe[class_var_string] == 0, class_var_string] = 88

        dataframe.loc[dataframe[class_var_string] == -99, class_var_string] = 0
        dataframe.loc[dataframe[class_var_string] == 88, class_var_string] = 1

        
    return dataframe
```

# perform Agglomerative Ward linkage Clustering 

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(linkage="ward", n_clusters=2)
clustering.fit(secondaryDf[['PC1','PC2']].to_numpy())



secondaryDf['AggClass'] = clustering.labels_


pp = sns.pairplot(data=secondaryDf,
                  y_vars=['PC2'],
                  x_vars=['PC1'],
                  kind = 'scatter', 
                 diag_kind='hist', corner = 'True', hue='AggClass', height=8,plot_kws={"s": 100}
                 )
leg_handles = pp._legend_data.values()
leg_labels = pp._legend_data.keys()
pp.fig.legend(handles=leg_handles, labels=leg_labels, loc='upper right', ncol=1)
pp._legend.remove() # remove existing legend
pp.set(xlabel='PC1',
       ylabel='PC2',
       title='AggWard (cluster centers = 2)')

#calculate and plot cluster centers at average
x_center = np.mean(secondaryDf[secondaryDf['AggClass'] == 0]['PC1'].to_numpy()),np.mean(secondaryDf[secondaryDf['AggClass'] == 1]['PC1'].to_numpy())
y_center = np.mean(secondaryDf[secondaryDf['AggClass'] == 0]['PC2'].to_numpy()),np.mean(secondaryDf[secondaryDf['AggClass'] == 1]['PC2'].to_numpy())



plt.scatter(x_center,y_center, marker='x',s=1000, facecolors='black')
plt.tight_layout()
# plt.savefig('../figures/pc_space_agg_clustered.jpg')
```
